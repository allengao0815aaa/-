```python
import requests
import re
import bs4
import json
import time
import random

headers = {
    "Host": "renshi.people.com.cn",
    "Connection": "keep-alive",
    "Pragma": "no-cache",
    "Cache-Control": "no-cache",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "Accept-Encoding": "gzip, deflate",
    "Accept-Language": "zh-CN,zh;q=0.9",
    "Cookie": "_ma_tk=d0nag8jst1bjjh7hehfh2bvb6olkw6co; sso_c=0; sfr=1; wdcid=6fbeb681e28521b7"
}
base_url = 'http://renshi.people.com.cn'
list_url = 'http://renshi.people.com.cn/index'
url_save_dir = './article_urls.txt'
json_save_dir = './content.json'
```

```python
# 爬取文章列表
article_urls = []
for i in range(1, 26):
    url = list_url + str(i)
    r = requests.get(url + '.html', headers = headers)
    print(url + '.html')
    if r.status_code == 200:
        r.encoding = r.apparent_encoding
        soup = bs4.BeautifulSoup(r.text, 'html.parser')
        articles = soup.select('div#p2Ab_1 > div.hdNews')
        for article in articles:
            article_urls.append(base_url + article.p.a.attrs['href'])
    else:
        print(r.status_code)
# 保存文章列表
with open(url_save_dir, 'w', encoding='utf8') as fp:
    for article_url in article_urls:
        fp.write(article_url + '\n')
```

```python
# 读取文章列表
article_urls = []
with open(url_save_dir, 'r', encoding='utf8') as fp:
    article_urls = fp.readlines()
for i in range(len(article_urls)):
    article_urls[i] = article_urls[i].rstrip('\n')
```

```python
# 提取文章列表中的文章
extracted = []
for i in range(len(article_urls)):
    r = requests.get(article_urls[i], headers = headers)
    print(article_urls[i])
    if r.status_code == 200:
        r.encoding = r.apparent_encoding
        soup = bs4.BeautifulSoup(r.text, 'html.parser')
        print(soup.title.text)
        passages = soup.select('div.show_text > p')
        content = []
        for passage in passages:
            if passage.text.find('更多报道') != -1:
                break
            content.append(passage.text)
        extracted.append({
            'title': soup.title.text,
            'content': content
        })
    else:
        print(r.status_code)
    num = random.randint(1,5)
    time.sleep(num)
```

```python
# 保存爬取的内容
with open(json_save_dir, 'w', encoding='utf8') as fp:
    json.dump(extracted, fp)
```

```python
# 读取爬取的内容
extracted = []
with open(json_save_dir, 'r', encoding='utf8') as fp:
    extracted = json.load(fp)
```

```python
# 提取简历
resumes = []
for article in extracted:
    find_resume = False
    resume = {}
    for i in range(len(article['content'])):
        article['content'][i] = re.sub('[\n\t]', '', article['content'][i])
        article['content'][i] = re.sub(',', '，', article['content'][i])
        article['content'][i] = re.sub(' ', '', article['content'][i])
        if article['content'][i].find('简历') != -1:
            if find_resume == True:
                print('WARNING: multiple resumes in an article.')
                resumes.append(resume)
                resume = {}
            find_resume = True
            resume['title'] = article['title']
            resume['name'] = article['content'][i]
            resume['content'] = []
            continue
        if find_resume == True:
            resume['content'].append(article['content'][i])
    if find_resume == True:
        resumes.append(resume)
```

```python
provinces = ["北京市", "天津市", "上海市", "重庆市", "内蒙古自治区", "广西壮族自治区", "西藏自治区", "宁夏回族自治区", "新疆维吾尔自治区", "河北省", "山西省", "辽宁省", "吉林省", "黑龙江省", "江苏省", "浙江省", "安徽省", "福建省", "江西省", "山东省", "河南省", "湖北省", "湖南省", "广东省", "海南省", "四川省", "贵州省", "云南省", "陕西省", "甘肃省", "青海省", "台湾省"]
for resume in resumes:
    resume['cut_name'] = ''
    resume['position'] = ''
    if resume['title'].find('任') != -1:
        resume['cut_name'] = resume['title'].split('任')[0]
        resume['position'] = resume['title'].split('任')[1]
    resume['gender'] = ''
    resume['birth'] = ''
    resume['ethnic'] = ''
    resume['party'] = ''
    resume['university'] = ''
    resume['PKU'] = ''
    resume['THU'] = ''
    resume['party_school'] = ''
    resume['birth_place'] = ''
    resume['degree'] = '0'
    for province in provinces:
        resume[province] = '0'
    for content in resume['content']:
        # 党派
        if content.find('共产党') != -1 or content.find('入党') != -1:
            resume['party'] = '共产党'
        # 大学相关
        if content.find('大学') != -1:
            resume['university'] = '1'
        # 党校相关
        if content.find('党校') != -1:
            resume['party_school'] = '1'
        # 北大相关
        if content.find('北京大学') != -1:
            resume['PKU'] = '1'
        # 清华相关
        if content.find('清华大学') != -1:
            resume['THU'] = '1'
        # 性别
        if content.find('男') != -1:
            resume['gender'] += '男'
        if content.find('女') != -1:
            resume['gender'] += '女'
        # 出生地
        if resume['birth_place'] == '':
            if content.find('人') != -1:
                resume['birth_place'] += content[content.rfind('，', 0, content.find('人')) + 1:content.find('人')]
        # 民族
        if resume['ethnic'] == '':
            if content.find('族') != -1:
                resume['ethnic'] += content[content.rfind('，', 0, content.find('族')) + 1:content.find('族') + 1]
        # 出生
        if content.find('出生') != -1:
            resume['birth'] += content[content.rfind('，', 0, content.find('出生')) + 1:content.find('出生')]
        if content.find('月生') != -1:
            resume['birth'] += content[content.rfind('，', 0, content.find('月生')) + 1:content.find('月生') + 1]
        # 学历
        if content.find('博士') != -1:
            resume['degree'] = '6'
        if content.find('硕士') != -1 and int(resume['degree']) < 5:
            resume['degree'] = '5'
        if content.find('研究生学历') != -1 and int(resume['degree']) < 5:
            resume['degree'] = '5'
        if content.find('本科') != -1 and int(resume['degree']) < 4:
            resume['degree'] = '4'
        if content.find('大学') != -1 and int(resume['degree']) < 4:
            resume['degree'] = '4'
        # 省份相关
        for province in provinces:
            if content.find(province) != -1:
                resume[province] = '1'
```

```python
with open('result.csv', 'w', encoding='utf8') as fp:
    fp.write('原始内容,姓名,性别,出生,是否共产党,职位,学历,北大,清华,大学,民族,籍贯,北京市,天津市,上海市,重庆市,内蒙古自治区,广西壮族自治区,西藏自治区,宁夏回族自治区,新疆维吾尔自治区,河北省,山西省,辽宁省,吉林省,黑龙江省,江苏省,浙江省,安徽省,福建省,江西省,山东省,河南省,湖北省,湖南省,广东省,海南省,四川省,贵州省,云南省,陕西省,甘肃省,青海省,台湾省' + '\n')
    for resume in resumes:
        for content in resume['content']:
            fp.write(content)
        fp.write(',' + resume['cut_name'])
        fp.write(',' + resume['gender'])
        fp.write(',' + resume['birth'])
        fp.write(',' + resume['party'])
        fp.write(',' + resume['position'])
        fp.write(',' + resume['degree'])
        fp.write(',' + resume['PKU'])
        fp.write(',' + resume['THU'])
        fp.write(',' + resume['university'])
        fp.write(',' + resume['ethnic'])
        fp.write(',' + resume['birth_place'])
        for province in provinces:
            fp.write(',' + resume[province])
        fp.write('\n')
```
